```python
top_10 = pokedex.nlargest(10, 'total_points')
```

```python
plt.figure(figsize=(10, 6))  # Define o tamanho da figura
plt.bar(top_10['name'], top_10['total_points'], color='skyblue')  # Plota o gráfico de barras
plt.xlabel('Categorias')  # Define o rótulo do eixo x
plt.ylabel('Valores')  # Define o rótulo do eixo y
plt.title('Top 10 Pokemons com maior points')  # Define o título do gráfico
plt.xticks(rotation=45)  # Rotaciona os rótulos do eixo x para melhor legibilidade
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Adiciona linhas de grade no eixo y
plt.tight_layout()  # Ajusta o layout para melhor visualização
plt.show()  # Mostra o gráfico
```

```python
plt.figure(figsize=(10, 6))  # Define o tamanho da figura
plt.fill_between(pokedex['generation'], pokedex['pokedex_number'], color='skyblue', alpha=0.4)  # Plota o gráfico de área
plt.plot(pokedex['generation'], pokedex['pokedex_number'], color='skyblue', alpha=0.6, linewidth=2)  # Adiciona as linhas do gráfico
plt.xlabel('Geração')  # Define o rótulo do eixo x
plt.ylabel('Pokemons')  # Define o rótulo do eixo y
plt.title('Numero de Pokemons por Geração')  # Define o título do gráfico
plt.xticks(rotation=45)  # Rotaciona os rótulos do eixo x para melhor legibilidade
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Adiciona linhas de grade no eixo y


plt.tight_layout()  
plt.show()  


```
